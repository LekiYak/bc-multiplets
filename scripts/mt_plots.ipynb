```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pygmt
import os
```

```python
def read_cmt(filepath):
    # read GCMT fms file for psmeca
    df = pd.read_csv(filepath, sep='\s+', header=0,
                             names=['longitude', 'latitude', 'depth', 'mrr', 'mtt', 'mff', 'mrt', 'mrf', 'mtf', 'exponent', 'event_name'])
    
    # convert eventid (event_name for psmeca plotting specs) to string
    df['event_name'] = df['event_name'].astype(str)

    return df
```

```python
df_2008 = read_cmt('/local/lyakuden/offshore_bc/bc-multiplets/gcmt/mt_comparison.txt')
```

```python
for index, row in df_2008.iterrows():

    fms_df = pd.DataFrame([row])

    event_name = row['event_name']
    
    fms_df = fms_df.drop(columns=['event_name'])

    fig = pygmt.Figure()
    pygmt.config(
        MAP_FRAME_TYPE="plain",
        MAP_FRAME_PEN="0p,white@100",
        MAP_TICK_PEN="0p,white@100",
    )
    fig.basemap(region=[-1,1,-1,1], projection="X3c/3c", frame="f")

    fig.meca(fms_df, scale="2c")

    # fig.show()
    
    fig.savefig(f"../figures/mt_plots/{event_name}.png", dpi=300, transparent=True)
```
